```python
# ---------------------------------------------------------
# Step 2: Define Project Paths
# ---------------------------------------------------------
# BASE_DIR is the main project folder inside Google Drive.
# RAW_DATA_DIR is where the original downloaded dataset will be stored.

import os

BASE_DIR = "/content/drive/MyDrive/UPI-Shield-AI-Transaction-Risk-Scoring"
RAW_DATA_DIR = os.path.join(BASE_DIR, "data", "raw")
PROCESSED_DATA_DIR = os.path.join(BASE_DIR, "data", "processed")

os.makedirs(RAW_DATA_DIR, exist_ok=True)
os.makedirs(PROCESSED_DATA_DIR, exist_ok=True)

print("Base Directory:", BASE_DIR)
print("Raw Data Directory:", RAW_DATA_DIR)
print("Processed Data Directory:", PROCESSED_DATA_DIR)
```

```python
# ---------------------------------------------------------
# Step 4: Configure Kaggle API Credentials
# ---------------------------------------------------------
# Kaggle expects kaggle.json to be stored inside /root/.kaggle/
# with secure file permissions.

import shutil

KAGGLE_DIR = "/root/.kaggle"
os.makedirs(KAGGLE_DIR, exist_ok=True)

source_path = "kaggle.json"
destination_path = os.path.join(KAGGLE_DIR, "kaggle.json")

shutil.copy(source_path, destination_path)
os.chmod(destination_path, 600)

print("Kaggle API credentials configured successfully.")
```

```python
# ---------------------------------------------------------
# Step 7: Verify Downloaded Files
# ---------------------------------------------------------
# This cell checks all files available inside data/raw.

raw_files = os.listdir(RAW_DATA_DIR)

print("Files inside data/raw:")
for file in raw_files:
    print("-", file)
```

```python
# ---------------------------------------------------------
# Step 8: Rename Dataset File
# ---------------------------------------------------------
# The original dataset filename is long, so we rename it to
# onlinefraud.csv for easier usage in all notebooks.

old_file = os.path.join(RAW_DATA_DIR, "PS_20174392719_1491204439457_log.csv")
new_file = os.path.join(RAW_DATA_DIR, "onlinefraud.csv")

if os.path.exists(old_file):
    os.rename(old_file, new_file)
    print("Dataset renamed successfully to onlinefraud.csv")
elif os.path.exists(new_file):
    print("onlinefraud.csv already exists. No renaming needed.")
else:
    print("Original dataset file not found. Please check downloaded file name.")

print("\nFinal files inside data/raw:")
for file in os.listdir(RAW_DATA_DIR):
    print("-", file)
```

```python
# ---------------------------------------------------------
# Step 9: Load Dataset
# ---------------------------------------------------------
# We load the dataset using pandas to verify that the CSV file
# is correctly downloaded, renamed, and readable.

import pandas as pd

DATA_PATH = os.path.join(RAW_DATA_DIR, "onlinefraud.csv")

df = pd.read_csv(DATA_PATH)

print("Dataset loaded successfully!")
print("Dataset shape:", df.shape)

df.head()
```

```python
# ---------------------------------------------------------
# Step 10: Check Dataset Columns
# ---------------------------------------------------------
# This helps us confirm that all required transaction columns
# are available for our project.

df.columns
```

```python
# ---------------------------------------------------------
# Step 11: Quick Dataset Verification
# ---------------------------------------------------------
# This cell gives a quick overview of dataset size, columns,
# missing values, and fraud class distribution.

print("Shape of dataset:", df.shape)

print("\nColumn names:")
print(df.columns.tolist())

print("\nMissing values:")
print(df.isnull().sum())

print("\nFraud class distribution:")
print(df["isFraud"].value_counts())

print("\nFraud class percentage:")
print(df["isFraud"].value_counts(normalize=True) * 100)
```
