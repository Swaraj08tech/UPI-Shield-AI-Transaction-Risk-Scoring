## Dataset Column Description

The dataset contains online/mobile money transaction records.

| Column Name | Description |
|---|---|
| `step` | Time step of transaction. In this dataset, 1 step represents 1 hour. |
| `type` | Type of transaction such as PAYMENT, TRANSFER, CASH_OUT, CASH_IN, DEBIT. |
| `amount` | Amount involved in the transaction. |
| `nameOrig` | Customer ID of the sender/origin account. |
| `oldbalanceOrg` | Sender's balance before the transaction. |
| `newbalanceOrig` | Sender's balance after the transaction. |
| `nameDest` | Customer ID of the receiver/destination account. |
| `oldbalanceDest` | Receiver's balance before the transaction. |
| `newbalanceDest` | Receiver's balance after the transaction. |
| `isFraud` | Target column. 1 means fraudulent transaction, 0 means normal transaction. |
| `isFlaggedFraud` | Rule-based fraud flag provided in the dataset. |

```python
# ---------------------------------------------------------
# Step 8: Check Column Names
# ---------------------------------------------------------
# This helps us verify all available columns in the dataset.

print("Total number of columns:", len(df.columns))
print("\nColumn names:")
for col in df.columns:
    print("-", col)
```

```python
# ---------------------------------------------------------
# Step 9: Basic Dataset Information
# ---------------------------------------------------------
# df.info() gives information about:
# 1. Number of non-null values
# 2. Data types
# 3. Memory usage

df.info()
```

```python
# ---------------------------------------------------------
# Step 10: Create Data Type Summary Table
# ---------------------------------------------------------
# This creates a clean summary table showing each column,
# its data type, number of unique values, and missing values.

data_summary = pd.DataFrame({
    "Column": df.columns,
    "Data Type": df.dtypes.values,
    "Unique Values": df.nunique().values,
    "Missing Values": df.isnull().sum().values,
    "Missing Percentage": (df.isnull().sum().values / len(df)) * 100
})

data_summary
```

```python
# ---------------------------------------------------------
# Step 11: Check Dataset Size and Memory Usage
# ---------------------------------------------------------
# Since this is a large dataset, we check memory usage.
# This helps us understand whether optimization may be needed later.

memory_usage_mb = df.memory_usage(deep=True).sum() / (1024 ** 2)

print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])
print("Memory usage: {:.2f} MB".format(memory_usage_mb))
```

```python
# ---------------------------------------------------------
# Step 12: Check Missing Values
# ---------------------------------------------------------
# Missing values can affect model training.
# Here, we check missing values column-wise.

missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100

missing_summary = pd.DataFrame({
    "Missing Values": missing_values,
    "Missing Percentage": missing_percentage
})

missing_summary
```

```python
# ---------------------------------------------------------
# Step 13: Missing Value Result
# ---------------------------------------------------------
# This gives a simple message based on missing value count.

total_missing = df.isnull().sum().sum()

if total_missing == 0:
    print("No missing values found in the dataset.")
else:
    print("Total missing values found:", total_missing)
```

```python
# ---------------------------------------------------------
# Step 14: Check Duplicate Records
# ---------------------------------------------------------
# Duplicate rows can create bias during analysis and training.
# We check the number and percentage of duplicate records.

duplicate_count = df.duplicated().sum()
duplicate_percentage = (duplicate_count / len(df)) * 100

print("Duplicate records:", duplicate_count)
print("Duplicate percentage: {:.4f}%".format(duplicate_percentage))
```

```python
# ---------------------------------------------------------
# Step 15: Statistical Summary of Numerical Columns
# ---------------------------------------------------------
# This gives count, mean, standard deviation, minimum,
# maximum, and quartile values for numerical columns.

df.describe()
```

```python
# ---------------------------------------------------------
# Step 16: Summary of Categorical Columns
# ---------------------------------------------------------
# The transaction type column is categorical.
# This cell summarizes all object/category columns.

categorical_columns = df.select_dtypes(include=["object"]).columns.tolist()

print("Categorical columns:", categorical_columns)

df[categorical_columns].describe()
```

```python
# ---------------------------------------------------------
# Step 17: Transaction Type Distribution
# ---------------------------------------------------------
# This shows how many transactions belong to each transaction type.

transaction_type_counts = df["type"].value_counts()

print("Transaction Type Distribution:")
print(transaction_type_counts)
```

```python
# ---------------------------------------------------------
# Step 18: Transaction Type Percentage
# ---------------------------------------------------------
# This shows transaction type distribution in percentage form.

transaction_type_percentage = df["type"].value_counts(normalize=True) * 100

transaction_type_summary = pd.DataFrame({
    "Transaction Count": transaction_type_counts,
    "Percentage": transaction_type_percentage
})

transaction_type_summary
```

```python
# ---------------------------------------------------------
# Step 19: Target Variable Distribution
# ---------------------------------------------------------
# isFraud is our target column.
# 0 means normal transaction.
# 1 means fraud transaction.

fraud_counts = df["isFraud"].value_counts()
fraud_percentage = df["isFraud"].value_counts(normalize=True) * 100

fraud_summary = pd.DataFrame({
    "Transaction Type": ["Normal", "Fraud"],
    "Count": [fraud_counts.get(0, 0), fraud_counts.get(1, 0)],
    "Percentage": [fraud_percentage.get(0, 0), fraud_percentage.get(1, 0)]
})

fraud_summary
```

```python
# ---------------------------------------------------------
# Step 20: Print Fraud Imbalance Clearly
# ---------------------------------------------------------
# This gives a clear class imbalance summary.
# Fraud detection datasets usually have very few fraud cases.

normal_count = fraud_counts.get(0, 0)
fraud_count = fraud_counts.get(1, 0)

print("Normal Transactions:", normal_count)
print("Fraud Transactions:", fraud_count)

print("\nNormal Transaction Percentage: {:.4f}%".format((normal_count / len(df)) * 100))
print("Fraud Transaction Percentage: {:.4f}%".format((fraud_count / len(df)) * 100))

imbalance_ratio = normal_count / fraud_count
print("\nClass Imbalance Ratio:")
print("For every 1 fraud transaction, there are approximately {:.0f} normal transactions.".format(imbalance_ratio))
```

## Initial Observation: Class Imbalance

The dataset is highly imbalanced because fraud transactions are much fewer than normal transactions.

This is important because a model can achieve very high accuracy by predicting most transactions as normal, but such a model may fail to detect actual fraud cases.

Therefore, during model evaluation, we should focus more on:

- Precision
- Recall
- F1-score
- ROC-AUC
- Confusion Matrix

For fraud detection, recall is especially important because missing an actual fraud transaction can be costly.

```python
# ---------------------------------------------------------
# Step 21: Fraud Distribution by Transaction Type
# ---------------------------------------------------------
# This table shows how many fraud transactions are present
# in each transaction type.

fraud_by_type = df.groupby("type")["isFraud"].agg(["count", "sum"])

fraud_by_type = fraud_by_type.rename(columns={
    "count": "Total Transactions",
    "sum": "Fraud Transactions"
})

fraud_by_type["Fraud Percentage"] = (
    fraud_by_type["Fraud Transactions"] / fraud_by_type["Total Transactions"]
) * 100

fraud_by_type = fraud_by_type.sort_values(by="Fraud Transactions", ascending=False)

fraud_by_type
```

```python
# ---------------------------------------------------------
# Step 22: Analyze isFlaggedFraud Column
# ---------------------------------------------------------
# isFlaggedFraud is a rule-based flag in the dataset.
# We check how many transactions were flagged by this rule.

flagged_counts = df["isFlaggedFraud"].value_counts()

print("isFlaggedFraud distribution:")
print(flagged_counts)

print("\nPercentage:")
print(df["isFlaggedFraud"].value_counts(normalize=True) * 100)
```

```python
# ---------------------------------------------------------
# Step 23: Compare isFraud and isFlaggedFraud
# ---------------------------------------------------------
# This helps us understand whether the rule-based flag catches
# actual fraud transactions properly.

fraud_flag_comparison = pd.crosstab(
    df["isFraud"],
    df["isFlaggedFraud"],
    rownames=["Actual Fraud"],
    colnames=["Flagged Fraud"]
)

fraud_flag_comparison
```

## Observation: `isFlaggedFraud`

The column `isFlaggedFraud` is a rule-based fraud flag already present in the dataset.

However, our goal is to build a machine learning-based risk scoring system. Therefore, we should be careful while using this column during model training.

If `isFlaggedFraud` directly leaks fraud-related rule information, it may make the model look better than it actually is.

We will analyze it further before deciding whether to use or remove it in model training.

```python
# ---------------------------------------------------------
# Step 24: Amount Summary for Normal vs Fraud Transactions
# ---------------------------------------------------------
# This compares transaction amount statistics for normal and
# fraud transactions.

amount_summary = df.groupby("isFraud")["amount"].describe()

amount_summary
```

```python
# ---------------------------------------------------------
# Step 25: Improve Amount Summary Readability
# ---------------------------------------------------------
# We rename target labels from 0/1 to Normal/Fraud for better
# readability.

amount_summary_readable = amount_summary.copy()
amount_summary_readable.index = ["Normal Transactions", "Fraud Transactions"]

amount_summary_readable
```

```python
# ---------------------------------------------------------
# Step 26: Balance Columns Summary
# ---------------------------------------------------------
# This gives statistical summary of sender and receiver balance
# columns.

balance_columns = [
    "oldbalanceOrg",
    "newbalanceOrig",
    "oldbalanceDest",
    "newbalanceDest"
]

df[balance_columns].describe()
```

```python
# ---------------------------------------------------------
# Step 27: Check Zero Balance After Transaction
# ---------------------------------------------------------
# In fraud transactions, sender balance becoming zero after
# transaction can be an important pattern.
# We check this behavior here.

zero_balance_summary = df.groupby("isFraud")["newbalanceOrig"].apply(lambda x: (x == 0).sum())

zero_balance_percentage = df.groupby("isFraud")["newbalanceOrig"].apply(lambda x: ((x == 0).sum() / len(x)) * 100)

zero_balance_result = pd.DataFrame({
    "Zero New Sender Balance Count": zero_balance_summary,
    "Zero New Sender Balance Percentage": zero_balance_percentage
})

zero_balance_result.index = ["Normal Transactions", "Fraud Transactions"]

zero_balance_result
```

```python
# ---------------------------------------------------------
# Step 28: Check Unique Sender and Receiver IDs
# ---------------------------------------------------------
# nameOrig and nameDest are customer/account identifiers.
# We check their uniqueness to understand whether they can
# be useful directly or should be removed later.

unique_sender_count = df["nameOrig"].nunique()
unique_receiver_count = df["nameDest"].nunique()

print("Unique sender IDs:", unique_sender_count)
print("Unique receiver IDs:", unique_receiver_count)

print("\nTotal transactions:", len(df))

print("\nSender uniqueness percentage: {:.2f}%".format((unique_sender_count / len(df)) * 100))
print("Receiver uniqueness percentage: {:.2f}%".format((unique_receiver_count / len(df)) * 100))
```

## Observation: Sender and Receiver ID Columns

The columns `nameOrig` and `nameDest` represent sender and receiver account/customer IDs.

These columns usually have very high cardinality, meaning they contain many unique values.

For a basic machine learning pipeline, these ID columns may not be directly useful because they can make the model memorize specific account IDs instead of learning transaction behavior.

Therefore, in the feature engineering notebook, we will mostly focus on transaction behavior features such as amount, transaction type, and balance changes.

```python
# ---------------------------------------------------------
# Step 29: Check Time Step Range
# ---------------------------------------------------------
# The step column represents time.
# We check the minimum and maximum step values.

print("Minimum step:", df["step"].min())
print("Maximum step:", df["step"].max())
print("Number of unique steps:", df["step"].nunique())
```

```python
# ---------------------------------------------------------
# Step 30: Create Dataset Quality Summary
# ---------------------------------------------------------
# This function summarizes important dataset quality details
# in one place.

def dataset_quality_summary(data):
    summary = {
        "Total Rows": data.shape[0],
        "Total Columns": data.shape[1],
        "Total Missing Values": data.isnull().sum().sum(),
        "Total Duplicate Rows": data.duplicated().sum(),
        "Fraud Transactions": data["isFraud"].sum(),
        "Normal Transactions": (data["isFraud"] == 0).sum(),
        "Fraud Percentage": (data["isFraud"].mean()) * 100,
        "Unique Transaction Types": data["type"].nunique()
    }

    return pd.DataFrame(summary.items(), columns=["Metric", "Value"])

quality_summary = dataset_quality_summary(df)
quality_summary
```
